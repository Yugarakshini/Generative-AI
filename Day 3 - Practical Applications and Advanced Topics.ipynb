# Cohere Chatbot

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import cohere
import webbrowser
from threading import Timer

# Initialize Cohere client with API key
co = cohere.Client(cohere_api_key)

# Function to get Cohere chatbot response
def get_cohere_response(user_input, chat_history):
    prompt = ""
    
    # Add chat history to the prompt
    for chat in chat_history:
        prompt += f"User: {chat['user']}\nBot: {chat['bot']}\n"

    prompt += f"User: {user_input}\nBot:"

    # Get the Cohere response
    response = co.generate(
        model='command-xlarge',
        prompt=prompt,
        max_tokens=100,
        temperature=0.7
    )

    return response.generations[0].text.strip()

# Initialize Dash app
app = dash.Dash(__name__)

# App layout with dark theme
app.layout = html.Div(style={'backgroundColor': '#1e1e1e', 'color': 'white', 'padding': '10px'}, children=[
    html.H1("Cohere Chatbot", style={'textAlign': 'center', 'color': '#FFFFFF'}),
    
    # Chat history display
    dcc.Textarea(
        id='chat-history',
        value='Welcome to the chatbot! Ask me anything.',
        style={'width': '100%', 'height': 300, 'backgroundColor': '#333333', 'color': 'white', 'border': '1px solid #444', 'padding': '10px'},
        readOnly=True
    ),
    
    # User input
    dcc.Input(
        id='user-input',
        type='text',
        placeholder='Type your message...',
        style={'width': '90%', 'padding': '10px', 'backgroundColor': '#444444', 'color': 'white', 'border': '1px solid #555'}
    ),
    
    # Send button
    html.Button('Send', id='send-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#28a745', 'color': 'white'}),
    
    # Clear button
    html.Button('Clear', id='clear-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#dc3545', 'color': 'white'}),
    
    # Hidden Div for storing chat history
    dcc.Store(id='chat-store', data=[])
])

# Callback to update chat and interact with Cohere
@app.callback(
    [Output('chat-history', 'value'), Output('chat-store', 'data'), Output('user-input', 'value')],
    [Input('send-btn', 'n_clicks'), Input('clear-btn', 'n_clicks')],
    [State('user-input', 'value'), State('chat-store', 'data')]
)
def update_chat(n_clicks_send, n_clicks_clear, user_input, chat_history):
    if n_clicks_clear > 0:
        # Clear the chat
        return 'Welcome to the chatbot! Ask me anything.', [], ''
    
    if n_clicks_send > 0 and user_input:
        # Get the bot response
        bot_response = get_cohere_response(user_input, chat_history)
        
        # Update chat history with user input and bot response
        chat_history.append({"user": user_input, "bot": bot_response})
        
        # Format chat for display
        chat_display = ""
        for chat in chat_history:
            chat_display += f"User: {chat['user']}\nBot: {chat['bot']}\n\n"
        
        # Clear user input field after sending
        return chat_display, chat_history, ''
    
    return dash.no_update, dash.no_update, dash.no_update

# Function to open the browser with the correct port
def open_browser(port):
    webbrowser.open_new(f'http://127.0.0.1:{port}/')

# Run the app
if __name__ == '__main__':
    port = 8051  # Set the port number
    Timer(1, open_browser, [port]).start()  # Pass the port to the open_browser function
    app.run_server(debug=False, use_reloader=False, port=port)

```

# Interview Trainer

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import cohere
import webbrowser
from threading import Timer

# Initialize Cohere client with API key
#cohere_api_key = 'your-cohere-api-key'  # Replace with your actual Cohere API key
co = cohere.Client(cohere_api_key)

# Function to get Cohere response for answers
def get_cohere_response(prompt):
    response = co.generate(
        model='command-xlarge',
        prompt=prompt,
        max_tokens=150,
        temperature=0.7
    )
    return response.generations[0].text.strip()

# Function to generate interview questions and answers based on job role
def generate_interview_questions_and_answers(job_role):
    # Generate a list of common interview questions
    questions_prompt = f"Generate a list of common interview questions for a {job_role} role."
    questions_response = get_cohere_response(questions_prompt)
    
    # Generate suggested answers for each question
    answers_prompt = f"Provide suggested answers for the following interview questions for a {job_role} role:\n{questions_response}"
    answers_response = get_cohere_response(answers_prompt)
    
    return questions_response, answers_response

# Initialize Dash app
app = dash.Dash(__name__)

# App layout with colorful theme
app.layout = html.Div(style={'backgroundColor': '#f0f8ff', 'color': '#333333', 'padding': '10px'}, children=[
    html.H1("Interview Training Chatbot", style={'textAlign': 'center', 'color': '#4a90e2'}),
    
    # Chat history display
    dcc.Textarea(
        id='chat-history',
        value='Welcome to the interview training chatbot! Please tell me your job role to start.',
        style={'width': '100%', 'height': 300, 'backgroundColor': '#e0f7fa', 'color': '#333333', 'border': '1px solid #4a90e2', 'padding': '10px'},
        readOnly=True
    ),
    
    # User input
    dcc.Input(
        id='user-input',
        type='text',
        placeholder='Type your message...',
        style={'width': '85%', 'padding': '10px', 'backgroundColor': '#b3e5fc', 'color': '#333333', 'border': '1px solid #4a90e2'}
    ),
    
    # Send button
    html.Button('Send', id='send-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#4caf50', 'color': 'white', 'border': 'none', 'padding': '10px', 'borderRadius': '5px'}),
    
    # Clear button
    html.Button('Clear', id='clear-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#f44336', 'color': 'white', 'border': 'none', 'padding': '10px', 'borderRadius': '5px'}),
    
    # Hidden Div for storing chat history and job role
    dcc.Store(id='chat-store', data={'history': [], 'job_role': None}),
    
    # Feedback and evaluation area
    html.Div(id='feedback', style={'marginTop': '20px', 'color': '#4a90e2'}),
])

# Callback to update chat and interact with Cohere
@app.callback(
    [Output('chat-history', 'value'), Output('chat-store', 'data'), Output('feedback', 'children')],
    [Input('send-btn', 'n_clicks'), Input('clear-btn', 'n_clicks')],
    [State('user-input', 'value'), State('chat-store', 'data')]
)
def update_chat(n_clicks_send, n_clicks_clear, user_input, chat_data):
    chat_history = chat_data['history']
    job_role = chat_data['job_role']

    if n_clicks_clear > 0:
        return 'Welcome to the interview training chatbot! Please tell me your job role to start.', {'history': [], 'job_role': None}, ''
    
    if n_clicks_send > 0 and user_input:
        if not job_role:
            # Set the job role
            job_role = user_input
            chat_data['job_role'] = job_role
            questions, answers = generate_interview_questions_and_answers(job_role)
            bot_response = f"Great! You are applying for a {job_role} role. Here are some common questions and suggested answers for this role:\n\nQuestions:\n{questions}\n\nSuggested Answers:\n{answers}"
            chat_history.append({"user": user_input, "bot": bot_response})
        else:
            # Generate and respond to interview questions based on the role
            bot_response = get_cohere_response(user_input)
            chat_history.append({"user": user_input, "bot": bot_response})
            
        # Format chat for display
        chat_display = ""
        for chat in chat_history:
            chat_display += f"User: {chat['user']}\nBot: {chat['bot']}\n\n"
        
        return chat_display, chat_data, ''
    
    return dash.no_update, dash.no_update, dash.no_update

# Function to open the browser with the correct port
def open_browser(port):
    webbrowser.open_new(f'http://127.0.0.1:{port}/')

# Run the app
if __name__ == '__main__':
    port = 8053  # Set the port number
    Timer(1, open_browser, [port]).start()  # Pass the port to the open_browser function
    app.run_server(debug=False, use_reloader=False, port=port)

```

# Scenario Generator & Role playing

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output, State
import cohere
import webbrowser
from threading import Timer

# Initialize Cohere client with API key

co = cohere.Client(cohere_api_key)

# Function to generate scenarios based on user input
def generate_scenario(role):
    prompt = f"Generate a realistic scenario for a person in the role of {role}. The scenario should be detailed and include challenges that the person might face."

    response = co.generate(
        model='command-xlarge',
        prompt=prompt,
        max_tokens=150,
        temperature=0.7
    )

    return response.generations[0].text.strip()

# Function to get AI response based on user input and scenario
def get_ai_response(user_input, scenario):
    prompt = f"Scenario: {scenario}\n\nUser: {user_input}\nAI:"

    response = co.generate(
        model='command-xlarge',
        prompt=prompt,
        max_tokens=100,
        temperature=0.7
    )

    return response.generations[0].text.strip()

# Initialize Dash app
app = dash.Dash(__name__)

# App layout with colorful theme
app.layout = html.Div(style={'backgroundColor': '#f9f9f9', 'color': '#333333', 'padding': '20px'}, children=[
    html.H1("Interactive Scenario-Based Training", style={'textAlign': 'center', 'color': '#FF5733', 'fontFamily': 'Arial'}),
    
    # Scenario display
    dcc.Textarea(
        id='scenario-text',
        value='Please enter your role to start generating a scenario.',
        style={'width': '100%', 'height': 250, 'backgroundColor': '#D1C4E9', 'color': '#1A237E', 'border': '2px solid #7E57C2', 'padding': '10px', 'borderRadius': '5px'},
        readOnly=True
    ),
    
    # User input
    dcc.Input(
        id='user-input',
        type='text',
        placeholder='Type your response...',
        style={'width': '80%', 'padding': '12px', 'backgroundColor': '#BBDEFB', 'color': '#0D47A1', 'border': '2px solid #2196F3', 'borderRadius': '5px'}
    ),
    
    # Send button
    html.Button('Send', id='send-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#4CAF50', 'color': 'white', 'border': 'none', 'padding': '12px', 'borderRadius': '5px', 'fontSize': '16px'}),
    
    # Clear button
    html.Button('Clear', id='clear-btn', n_clicks=0, style={'margin': '10px', 'backgroundColor': '#F44336', 'color': 'white', 'border': 'none', 'padding': '12px', 'borderRadius': '5px', 'fontSize': '16px'}),
    
    # Feedback area
    html.Div(id='feedback', style={'marginTop': '20px', 'color': '#FF5722', 'fontSize': '18px', 'fontWeight': 'bold'}),
    
    # Hidden Div for storing scenario and role
    dcc.Store(id='scenario-store', data={'scenario': None, 'role': None}),
])

# Callback to update scenario and interact with Cohere
@app.callback(
    [Output('scenario-text', 'value'), Output('feedback', 'children')],
    [Input('send-btn', 'n_clicks'), Input('clear-btn', 'n_clicks')],
    [State('user-input', 'value'), State('scenario-store', 'data')]
)
def update_scenario(n_clicks_send, n_clicks_clear, user_input, store_data):
    role = store_data['role']
    scenario = store_data['scenario']

    if n_clicks_clear > 0:
        return 'Please enter your role to start generating a scenario.', ''
    
    if n_clicks_send > 0:
        if not role:
            # Set the role and generate a scenario
            role = user_input
            scenario = generate_scenario(role)
            store_data['role'] = role
            store_data['scenario'] = scenario
            return scenario, ''
        else:
            # Get AI response based on user input and scenario
            ai_response = get_ai_response(user_input, scenario)
            feedback = f"AI Response: {ai_response}"
            return scenario, feedback
    
    return dash.no_update, dash.no_update

# Function to open the browser with the correct port
def open_browser(port):
    webbrowser.open_new(f'http://127.0.0.1:{port}/')

# Run the app
if __name__ == '__main__':
    port = 8052  # Set the port number
    Timer(1, open_browser, [port]).start()  # Pass the port to the open_browser function
    app.run_server(debug=False, use_reloader=False, port=port)


```
